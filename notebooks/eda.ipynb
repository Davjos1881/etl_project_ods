```python
# Realize a EDA (Exploratory Data Analysis) on the dataset of candidates 
import pandas as pd
from tabulate import tabulate

# Define the path to the dataset
path = r'C:\Users\btigr\Documents\UAO\5\ETL\ETL_2026_1\proyecto_1\etl_project_ods-main\etl_project_ods-main\raw\incautaciones.csv'

# Load the dataset
df = pd.read_csv(path, sep=',')

# Show the first few rows of the dataset
print('Showing first few rows of the dataset:')
print(tabulate(df.head(), headers='keys', tablefmt='psql'))
```

```python
# Explore type of data in each column
print('\nData types of each column:')
print(tabulate(df.dtypes.to_frame(), headers='keys', tablefmt='psql'))

# Know how many rows and columns the dataset has
print('\nNumber of rows and columns in the dataset:')
print(tabulate([df.shape], headers=['Rows', 'Columns'], tablefmt='psql'))
```

```python
# Know nulls and duplicates
print('\nNumber of null values in each column:')
print(tabulate(df.isnull().sum().to_frame(), headers='keys', tablefmt='psql'))

print('\nNumber of duplicate rows in the dataset:')
print(df.duplicated().sum())
```

```python
# Conocer municipios donde hay mas incautaciones
print('\nMunicipios con mas incautaciones')
print(tabulate(df['Municipio'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))

# Bar plot horizontal 
import matplotlib.pyplot as plt

top_countries = df['Municipio'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_countries.index, top_countries.values)
plt.xlabel('Cantidad de incautaciones')
plt.title('Departamentos con mas incautaciones')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Conocer autoridades con mas incautaciones
print('\nAutoridades con mas incautaciones')
print(tabulate(df['Autoridad que incauto'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))

# Bar plot horizontal 
import matplotlib.pyplot as plt

top_countries = df['Autoridad que incauto'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_countries.index, top_countries.values)
plt.xlabel('Cantidad de incautaciones')
plt.title('Autoridades con mas incautaciones')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Conocer especies mas incautadas
print('\nEspecies mas incautadas')
print(tabulate(df['nom tipo especie'].value_counts().head(10).to_frame(), headers='keys', tablefmt='psql'))

# Bar plot horizontal 
import matplotlib.pyplot as plt

top_countries = df['nom tipo especie'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_countries.index, top_countries.values)
plt.xlabel('Cantidad de incautaciones')
plt.title('Especies mas incautadas')
plt.gca().invert_yaxis()
plt.show()
```
